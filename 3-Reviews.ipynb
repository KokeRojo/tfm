```python
df_reviews['review'] = df_reviews['review'].astype(str)
df_reviews.info()
```

```python
import re
import pandas as pd
import time

from transformers import pipeline

def analyze_sentiment_batch(classifier, texts, ids):
    max_tokens = 512
    truncated_texts = [str(text)[:max_tokens] for text in texts]
    
    results = classifier(truncated_texts)
    
    polarities = [int(re.search(r'\d+', result['label']).group()) for result in results]
    scores = [round(result['score'], 4) for result in results]
    
    return ids, polarities, scores

inicio = time.time()

classifier = pipeline('sentiment-analysis', model="nlptown/bert-base-multilingual-uncased-sentiment")
    
# Divide los datos en lotes de tamaño 10000
batch_size = 1000
batches = [(df_reviews['review'][i:i+batch_size].tolist(),
            df_reviews['id_review'][i:i+batch_size].tolist()) for i in range(0, len(df_reviews), batch_size)]

all_ids = []
all_polarities = []
all_scores = []

def process_batch_and_append(batch):
    print(len(batch))
    texts, ids = batch
    ids, polarities, scores = analyze_sentiment_batch(classifier, texts, ids)
    all_ids.extend(ids)
    all_polarities.extend(polarities)
    all_scores.extend(scores)

df_batches = pd.DataFrame(batches, columns=['texts', 'ids'])

df_batches.apply(process_batch_and_append, axis=1)

df_sentiment = pd.DataFrame({'id_review': all_ids, 'polarity': all_polarities, 'num_score': all_scores})

print(df_sentiment)

# Guarda el DataFrame en un archivo CSV
nombre_archivo = './data/sentimientos1.csv'
df_sentiment.to_csv(nombre_archivo, index=False, sep=',', decimal='.')

fin = time.time()
tiempo_transcurrido = fin - inicio
print("Tiempo transcurrido:", tiempo_transcurrido, "segundos")
```
